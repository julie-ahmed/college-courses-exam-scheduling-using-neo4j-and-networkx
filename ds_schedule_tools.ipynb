```python
def greedy_colouring_algorithm(courses_network, colour_list):
  nodes=list(courses_network.nodes())
  rnd.shuffle(nodes)
  coloured_nodes={}
  for node in nodes:
    node_neighbors=list(dict(courses_network[node]).keys())       #This view is an iterable (not a list), but it can be converted into a list.
    forbidden_colours_of_every_node=[]
    for this_neighbor in node_neighbors :
      if this_neighbor in coloured_nodes:
        forbidden_colours_of_every_node.append(coloured_nodes[this_neighbor])
    for colour in colour_list:
      if colour not in forbidden_colours_of_every_node:
        coloured_nodes[node]=colour
        break
  return coloured_nodes
```

```python
first_colouring_result=greedy_colouring_algorithm(courses_network=courses_network,colour_list=colour_list)
node_color_map = [first_colouring_result.get(node, 'white') for node in courses_network.nodes()]


nx.draw(courses_network, with_labels=False, node_color=node_color_map, font_weight='bold', node_size=80)
plt.show()
```

```python
def another_greedy_coloring_algorithm(network, colors):
    nodes = list(network.nodes())
    rnd.shuffle(nodes) # step 1 random ordering
    for node in nodes:
        dict_neighbors = dict(network[node])
# gives names of nodes that are neighbors
        nodes_neighbors = list(dict_neighbors.keys())

        forbidden_colors = []
        for neighbor in nodes_neighbors:
            if len(network.nodes.data()[neighbor].keys()) == 0:
                # if the neighbor has no color, proceed
                continue
            else:
                # if the neighbor has a color,
                # this color is forbidden
                forbidden_colors.append(network.nodes.data()[neighbor]['color'])
         # assign the first color
        # that is not forbidden
        for color in colors:
            # step 2: start everytime at the top of the colors,
            # so that the smallest number of colors is used
            if color in forbidden_colors:
                continue
            else:
                # step 3: color one node at the time
                network.nodes[node]['color'] = color
                break
another_greedy_coloring_algorithm(courses_network, colour_list)
colors_nodes = [data['color'] for v, data in courses_network.nodes(data=True)]
nx.draw(courses_network, node_color=colors_nodes, with_labels=False)
```

```python



from neo4j import GraphDatabase

class GraphToNeo4j:
    def __init__(self, uri, user, password):
        self.driver = GraphDatabase.driver(uri, auth=(user, password))

    def close(self):
        self.driver.close()

    def create_node(self, tx, node_id, color):
        tx.run("""
            MERGE (n:Node {id: $id})
            SET n.color = $color
        """, id=node_id, color=color)

    def create_edge(self, tx, source, target):
        tx.run("""
            MATCH (a:Node {id: $source})
            MATCH (b:Node {id: $target})
            MERGE (a)-[:CONNECTED_TO]->(b)
        """, source=source, target=target)

    def upload_graph(self, graph):
        with self.driver.session() as session:
            for node_id, attrs in graph.nodes(data=True):
                color = attrs.get('color', 'unknown')
                session.execute_write(self.create_node, node_id, color)

            for source, target in graph.edges():
                session.write_transaction(self.create_edge, source, target)

# Connect to Neo4j and upload
db = GraphToNeo4j("neo4j+s://clkdmvjf.databases.neo4j.io", "neo4j", "jjjjjjjjjwdenvjvjntbvudjmsmvnvj")
db.upload_graph(courses_network)
db.close()

```
